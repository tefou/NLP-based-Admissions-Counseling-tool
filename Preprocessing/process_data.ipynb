```python
def remove_punctuation(text):
    #a function for removing punctuation 
    import string
    #replacing the punctuations with no space
    #which in effect delates the punctuation marks
    translator = str.maketrans(list)
    #return the text stripped of punctution marks
    return text.translate(translator)
```

```python
df['Question'] = df['Question'].apply(remove_punctuation)
df.head(20)
```

```python
with open('stopword.txt', 'r', encoding='utf-8') as f:
    sw = f.readlines()
sw = [i[:-1] for i in sw]
sw
```

```python
def stopwords(text):
    #removing the stopwords and lowercasing the selected words
    text = [word.lower()for word in text.split() if word.lower() not in sw]
    #joining the list of words with space separator
    return " ".join(text)
```

```python
try:
    df['Question'] = df['Question'].apply(stopwords)
except:
    df.head
```

One Hot Encoding

```python
# Tạo DataFrame từ dữ liệu
data = pd.read_csv('Process_data.csv', header=0, delimiter=',', encoding='utf-8')

df = pd.DataFrame(data)

# Thực hiện one-hot encoding cho cột Label
one_hot_encoded = pd.get_dummies(df['Label'])
# Chuyển đổi từ True/False thành 0/1
binary_encoding = one_hot_encoded.astype(int)
# Ghép kết quả vào DataFrame gốc
df = pd.concat([df, binary_encoding], axis=1)
df
```

```python

# Thống kê dòng dữ liệu
row_count = data.shape[0]  # Số lượng dòng
print("Số lượng dòng dữ liệu:", row_count)

# Thống kê cột dữ liệu
column_count = data.shape[1]  # Số lượng cột
print("Số lượng cột dữ liệu:", column_count)
```

```python
# Thống kê theo nhóm câu hỏi
group_counts = data['Label'].value_counts()

# Vẽ biểu đồ cột
plt.bar(group_counts.index, group_counts.values)
plt.xlabel('Nhóm câu hỏi')
plt.ylabel('Số lượng')
plt.title('Thống kê số lượng câu hỏi theo nhóm')
plt.show()
```
